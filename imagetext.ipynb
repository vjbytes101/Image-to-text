```python
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.3 * r + 0.59 * g + 0.11 * b

    return gray
img= rgb2gray(img1)
lowValY = 200
```

```python
w, h = (int)(x_s/5), (int)(y_s/9);
l_t = img[:h,:w]
l_t = np.asarray(l_t)
low_values_flags = l_t < lowValY
l_t[low_values_flags] = 0 
plt.imshow(l_t,cmap="gray")
plt.show()
w,h
```
